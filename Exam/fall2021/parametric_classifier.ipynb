```python
grass = [68, 65, 67]
road = [70, 80, 75]
sky = [77, 92, 89]

min_hu = 69
max_hu = 88

hu_range = np.arange(min_hu, max_hu, 1.0)

pdf_grass = norm.pdf(hu_range, np.mean(grass), np.std(grass))
pdf_road = norm.pdf(hu_range, np.mean(road), np.std(road))
pdf_sky = norm.pdf(hu_range, np.mean(sky), np.std(sky))

plt.plot(hu_range, pdf_grass, 'g', label="bone")
plt.plot(hu_range, pdf_road, 'b', label="liver")
plt.plot(hu_range, pdf_sky, 'y', label="fat")
```
