```python
from scipy.ndimage import correlate
from skimage import io, color, morphology
from skimage.util import img_as_float, img_as_ubyte
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creating a small and simple image
input_img = np.arange(25).reshape(5,5)
input_img
```

```python
#Create a filter
weights = [[0,1,0],
           [1,2,1],
           [0,1,0]]
```

```python
res_img = correlate(input_img, weights)
res_img
```

```python
#Change mode of the correlate function
#Pad the edges with a constant value (cval)
res_img = correlate(input_img, weights, mode="constant", cval=10)
print(res_img)
#Pad the edges with a reflection of the pixels at the edge
res_img = correlate(input_img, weights, mode="reflect") #This is also the defualt mode
print(res_img)
```
